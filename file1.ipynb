```python
import kagglehub
import shutil
import os

# Download latest version
path = kagglehub.dataset_download("kevinkonstas/server-load-and-thermal-data")

print("Path to dataset files:", path)

# Copy to current repo
dest_dir = "./server-load-and-thermal-data"
if not os.path.exists(dest_dir):
    shutil.copytree(path, dest_dir)
    print(f"Data copied to {dest_dir}")
else:
    print(f"Data already exists at {dest_dir}")
```
