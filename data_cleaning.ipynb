#### Recalculate Physics With Independent Effects
- **Temperature**: depends on previous temp, CPU/Cooling ratio, ambient temp
- **Fan Speed**: depends on previous fan speed, temperature × fan health
- **Power**: base + CPU-proportional + rare anomaly spikes

```python
# ── Temperature Model ──
# Temp(t) = 0.85*Temp(t-1) + 0.1*(CPU(t)/Cooling_Efficiency) + 0.05*Ambient + noise
new_temps = []
for node, grp in df.groupby("Node_ID"):
    temps = grp["CPU1_Temp_C"].values.copy()
    cpu   = grp["CPU_Utilization_%"].values
    cool  = grp["Cooling_Efficiency"].values
    amb   = grp["Ambient_Temp_C"].values
    for i in range(1, len(temps)):
        temps[i] = (0.85 * temps[i-1]
                    + 0.1 * (cpu[i] / cool[i])
                    + 0.05 * amb[i]
                    + np.random.normal(0, 0.3))
    new_temps.append(pd.Series(temps, index=grp.index))

df["CPU1_Temp_C"] = pd.concat(new_temps)
df["CPU2_Temp_C"] = df["CPU1_Temp_C"] + np.random.normal(0.5, 0.3, len(df))
df["Exhaust_Temp_C"] = df["CPU1_Temp_C"] + 10 + np.random.normal(0, 1, len(df))

# ── Fan Speed Model ──
# Fan(t) = 0.8*Fan(t-1) + 0.2*(Temp(t-1) * Fan_Health)
new_fans = []
for node, grp in df.groupby("Node_ID"):
    fans  = grp["Avg_Fan_Speed_RPM"].values.copy()
    temps = grp["CPU1_Temp_C"].values
    fh    = grp["Fan_Health"].values
    for i in range(1, len(fans)):
        fans[i] = 0.8 * fans[i-1] + 0.2 * (temps[i-1] * fh[i])
    new_fans.append(pd.Series(fans, index=grp.index))

df["Avg_Fan_Speed_RPM"] = pd.concat(new_fans)

# ── Power Model with rare anomaly spikes ──
anomaly = np.where(
    np.random.random(len(df)) < 0.02,       # 2% chance of spike
    np.random.uniform(50, 150, len(df)),     # spike magnitude
    0
)
df["System_Power_W"] = (100
    + 2.5 * df["CPU_Utilization_%"]
    + anomaly
    + np.random.normal(0, 5, len(df))
)

print("✓ Physics recalculated with independent effects")
print(f"  CPU1_Temp_C  range: {df['CPU1_Temp_C'].min():.1f} – {df['CPU1_Temp_C'].max():.1f}")
print(f"  Fan RPM      range: {df['Avg_Fan_Speed_RPM'].min():.0f} – {df['Avg_Fan_Speed_RPM'].max():.0f}")
print(f"  Power (W)    range: {df['System_Power_W'].min():.0f} – {df['System_Power_W'].max():.0f}")
```

#### Target = Multi-Factor Overload (Cumulative Stress)
Instead of a simple CPU threshold, overload depends on **accumulated stress** across CPU, temperature, power, cooling efficiency, and fan health over the last 6 timesteps.

```python
# Instantaneous stress index (multi-factor)
df["Instant_Stress"] = (
    0.50 * (df["CPU_Utilization_%"] / 100)
    + 0.20 * (df["CPU1_Temp_C"] / 100)
    + 0.15 * (df["System_Power_W"] / 500)
    + 0.10 * (1 - df["Cooling_Efficiency"])
    + 0.05 * (1 - df["Fan_Health"])
)

print(f"Instant_Stress range: {df['Instant_Stress'].min():.3f} – {df['Instant_Stress'].max():.3f}")
print(f"Instant_Stress mean:  {df['Instant_Stress'].mean():.3f}")
```

```python
# Cumulative stress: rolling mean over last 6 timesteps (30 min window)
df["Cumulative_Stress"] = (
    df.groupby("Node_ID")["Instant_Stress"]
      .rolling(6)
      .mean()
      .reset_index(level=0, drop=True)
)

# Current overload flag based on accumulated stress
# Threshold 0.60 ≈ P80 → ~19% overload rate (realistic and ML-friendly)
df["Overload"] = (df["Cumulative_Stress"] > 0.60).astype(int)
print(f"Current Overload rate: {df['Overload'].mean():.3%}")
```

```python
# Target: will Overload occur 2 steps (10 min) into the future?
df["Overload_Future_10min"] = (
    df.groupby("Node_ID")["Overload"]
      .shift(-2)
)
```

```python
df = df.dropna(subset=["Overload_Future_10min"])
df["Overload_Future_10min"] = df["Overload_Future_10min"].astype(int)
```

```python
important_cols = [
    "CPU_Utilization_%",
    "CPU1_Temp_C",
    "CPU2_Temp_C",
    "System_Power_W",
    "Avg_Fan_Speed_RPM"
]
```

```python
for col in important_cols:
    for lag in [1, 2, 3]:
        df[f"{col}_lag{lag}"] = (
            df.groupby("Node_ID")[col]
              .shift(lag)
        )
```

```python
for col in important_cols:
    df[f"{col}_roll_mean_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).mean())
    )
```

```python
for col in important_cols:
    df[f"{col}_roll_std_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).std())
    )
```

```python
df = df.dropna()
```

```python
import numpy as np

## Split by Node_ID: train on 80% of nodes, test on unseen nodes
all_nodes = df["Node_ID"].unique()
np.random.seed(42)
np.random.shuffle(all_nodes)

split_idx = int(len(all_nodes) * 0.8)
train_nodes = all_nodes[:split_idx]
test_nodes = all_nodes[split_idx:]

print(f"Train nodes ({len(train_nodes)}): {sorted(train_nodes)}")
print(f"Test nodes ({len(test_nodes)}): {sorted(test_nodes)}")

train_df = df[df["Node_ID"].isin(train_nodes)]
test_df = df[df["Node_ID"].isin(test_nodes)]

drop_cols = [
    "Timestamp",
    "Node_ID",
    "Instant_Stress",
    "Cumulative_Stress",
    "Overload",
    "Overload_Future_10min"
]

X_train = train_df.drop(columns=drop_cols)
y_train = train_df["Overload_Future_10min"]

X_test = test_df.drop(columns=drop_cols)
y_test = test_df["Overload_Future_10min"]

print(f"\nTrain: {X_train.shape[0]} rows, {X_train.shape[1]} features")
print(f"Test:  {X_test.shape[0]} rows")
print(f"Features: {X_train.columns.tolist()}")
```
