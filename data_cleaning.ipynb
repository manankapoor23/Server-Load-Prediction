#### Target = High Load at T + 10 minutes

```python
HIGH_LOAD_THRESHOLD = 80
df["High_Load"] = (df["CPU_Utilization_%"] > HIGH_LOAD_THRESHOLD).astype(int)
```

```python
df["High_Load_Future_10min"] = (
    df.groupby("Node_ID")["High_Load"]
      .shift(-2)
)
```

```python
df = df.dropna(subset=["High_Load_Future_10min"])
```

```python
df["High_Load_Future_10min"] = df["High_Load_Future_10min"].astype(int)
```

```python
important_cols = [
    "CPU_Utilization_%",
    "CPU1_Temp_C",
    "CPU2_Temp_C",
    "System_Power_W",
    "Avg_Fan_Speed_RPM"
]
```

```python
for col in important_cols:
    for lag in [1, 2, 3]:
        df[f"{col}_lag{lag}"] = (
            df.groupby("Node_ID")[col]
              .shift(lag)
        )
```

```python
for col in important_cols:
    df[f"{col}_roll_mean_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).mean())
    )
```

```python
for col in important_cols:
    df[f"{col}_roll_std_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).std())
    )
```

```python
df = df.dropna()
```

```python
import numpy as np

## Split by Node_ID: train on 80% of nodes, test on unseen nodes
all_nodes = df["Node_ID"].unique()
np.random.seed(42)
np.random.shuffle(all_nodes)

split_idx = int(len(all_nodes) * 0.8)
train_nodes = all_nodes[:split_idx]
test_nodes = all_nodes[split_idx:]

print(f"Train nodes ({len(train_nodes)}): {sorted(train_nodes)}")
print(f"Test nodes ({len(test_nodes)}): {sorted(test_nodes)}")

train_df = df[df["Node_ID"].isin(train_nodes)]
test_df = df[df["Node_ID"].isin(test_nodes)]

drop_cols = [
    "Timestamp",
    "Node_ID",
    "High_Load",
    "High_Load_Future_10min"
]

X_train = train_df.drop(columns=drop_cols)
y_train = train_df["High_Load_Future_10min"]

X_test = test_df.drop(columns=drop_cols)
y_test = test_df["High_Load_Future_10min"]

print(f"\nTrain: {X_train.shape[0]} rows, Test: {X_test.shape[0]} rows")
```

```python
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()

scale_pos_weight = float(neg / pos)
print(scale_pos_weight)
```
