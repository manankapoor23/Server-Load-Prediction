```python
cols = df.columns.drop("Node_ID")
df[cols] = df.groupby("Node_ID")[cols].ffill()
```

```python
cols = df.columns.drop("Node_ID")
df[cols] = df.groupby("Node_ID")[cols].bfill()
```

```python
df = df[df["CPU_Utilization_%"] >= 0]
```

```python
df = df[df["CPU_Utilization_%"] <= 100]
```

```python
df = df.drop_duplicates()
```

```python
numeric_cols = df.columns.drop(["Timestamp"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
print(df["Timestamp"].dtype)
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
```

```python
df["hour"] = df["Timestamp"].dt.hour
df["day_of_week"] = df["Timestamp"].dt.dayofweek
```

#### Target = High Load at T + 30 minutes

```python
HIGH_LOAD_THRESHOLD = 80
df["High_Load"] = (df["CPU_Utilization_%"] > HIGH_LOAD_THRESHOLD).astype(int)
```

```python
df["High_Load_Future_30min"] = (
    df.groupby("Node_ID")["High_Load"]
      .shift(-6)
)
```

```python
df = df.dropna(subset=["High_Load_Future_30min"])
```

```python
df["High_Load_Future_30min"] = df["High_Load_Future_30min"].astype(int)
```

```python
print(df[["CPU_Utilization_%", 
          "High_Load", 
          "High_Load_Future_30min"]].head(150))
```

```python
print(df["High_Load_Future_30min"].value_counts(normalize=True))
```

```python
important_cols = [
    "CPU_Utilization_%",
    "CPU1_Temp_C",
    "CPU2_Temp_C",
    "System_Power_W",
    "Avg_Fan_Speed_RPM"
]
```

```python
for col in important_cols:
    for lag in [1, 2, 3]:
        df[f"{col}_lag{lag}"] = (
            df.groupby("Node_ID")[col]
              .shift(lag)
        )
```

```python
for col in important_cols:
    df[f"{col}_roll_mean_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).mean())
    )
```

```python
for col in important_cols:
    df[f"{col}_roll_std_3"] = (
        df.groupby("Node_ID")[col]
          .transform(lambda x: x.rolling(3).std())
    )
```

```python
df = df.dropna()
```

```python
drop_cols = [
    "Timestamp",
    "High_Load",
    "High_Load_Future_30min"
]

X = df.drop(columns=drop_cols)
y = df["High_Load_Future_30min"]
```

```python
split_index = int(len(df) * 0.8)

X_train = X.iloc[:split_index]
X_test  = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test  = y.iloc[split_index:]
```

```python
## Save the cleaned data
X_train.to_csv("/Users/manankapoor/Desktop/ai_proj/cleaned_data/X_train.csv", index=False)
X_test.to_csv("/Users/manankapoor/Desktop/ai_proj/cleaned_data/X_test.csv", index=False)
y_train.to_csv("/Users/manankapoor/Desktop/ai_proj/cleaned_data/y_train.csv", index=False)
y_test.to_csv("/Users/manankapoor/Desktop/ai_proj/cleaned_data/y_test.csv", index=False)    
```
